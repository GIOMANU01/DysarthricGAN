```python
import os
import wave
import contextlib
import pandas as pd
from pydub import AudioSegment
import matplotlib.pyplot as plt
import sys
print(sys.executable)


```

```python
dict_file = "C:/Users/gioel/Documents/MFA/pretrained_models/dictionary/english_mfa.dict"
excel_file = "C:/Users/gioel/Desktop/patient_GAN/patient_M04/lista_parole.xlsx"
```

```python

# Legge il file .dict
with open(dict_file, "r", encoding="utf-8") as f:
    dict_words = set()
    for line in f:
        if line.strip():
            word = line.split()[0]  # prende la prima colonna (la parola)
            dict_words.add(word.strip().lower())
```

```python
# Legge il file Excel
df = pd.read_excel(excel_file)
excel_words = set(df['WORD'].dropna().str.strip().str.lower())
```

```python
#Trova le parole che NON sono nel .dict
missing_words = sorted(excel_words - dict_words)
```

```python
print(f"Trovate {len(missing_words)} parole non presenti nel file dict:\n")
with open("C:/Users/gioel/Desktop/missing_words.txt", "w", encoding="utf-8") as out:
    out.write("\n".join(missing_words))
```
