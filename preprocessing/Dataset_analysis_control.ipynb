### Function to compute SNR

```python
import numpy as np
from scipy.io import wavfile
from scipy.signal import medfilt


def stima_rumore(segnale):
    # Applica un filtro mediano per stimare il rumore di fondo
    rumore_stimato = medfilt(segnale, kernel_size=3)
    
    # Calcola la differenza tra il segnale e l'estimatore del rumore
    rumore = segnale - rumore_stimato
    
    return rumore

def calcola_SNR(audio_file):
    # Carica il file audio
    FS, data = wavfile.read(audio_file)
    
    # Converti il segnale in float64
    segnale = data.astype(np.float64)
    
    # Stima del rumore
    rumore = stima_rumore(segnale)
    
    # Calcola l'SNR in dB
    SNR = 10 * np.log10(np.sum(segnale**2) / np.sum(rumore**2))
    
    return SNR
```

```python
df = pd.read_excel('C:/Users/gioel/Desktop/noisereduced-uaspeech-control/data.xlsx', )
df = df['FILE_PATH']
df.head()
SNR_value =  []
for elements in tqdm(df):
    try:
        valori = calcola_SNR(elements)
        SNR_value.append(valori)
    except:
        print(f'file errore: {elements}')
        SNR_value.append(0)
        
```

#### Add new colum to the dataframe

```python
df = pd.read_excel('C:/Users/gioel/Desktop/noisereduced-uaspeech-control/data.xlsx', )
df['SNR'] = SNR_value    
```
